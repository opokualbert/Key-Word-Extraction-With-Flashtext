# Keyword Extraction with Flashtext

```python
from flashtext import KeywordProcessor
```

```python
df_orig_1 = df_orig.loc[:,['Doc_Id','review_text']]
```

```python
keywords_names= ['dirty','mosquitoes','bed bugs','washroom']
```

```python
keyword_processor = KeywordProcessor()
```

```python
for keyword_name in keywords_names:
    keyword_processor.add_keyword(keyword_name)
    keywords_found = [keyword_processor.extract_keywords(word) for word in df_orig_1.review_text]
keywords_found
```

```python
df2 = pd.DataFrame(keywords_found)
df2.sample(5)
```

```python
df3 = df_orig_1.join(df2, how = 'outer')
df3.head()
```

```python
df4 = pd.melt(df3, id_vars=['Doc_Id','review_text'])
```

##### 

```python
df5= df4[pd.notnull(df4['value'])]
df5
```

```python
df6 = df5.loc[:,['Doc_Id','review_text','value']]
```

```python
df6.sort_values(by=['Doc_Id'])
```

```python
df7= df6.drop_duplicates(subset=['Doc_Id', 'value'],keep='last')
df8 = df7.sort_values(by=['Doc_Id'])
```
